```python
###4
#File path for storing data
Budget_File="budget.csv"
def initialize_csv():
  with open(Budget_File,mode='a',newline='' ) as file:
    if file.tell()==0:
     csv.writer(file).writerow(["Event Name","Venue Budget","Catering Budget","Entertainment Budget","Miscellaneous Budget"])

# Function to set the budget for an event
def set_budget():
    Event_Name = input("Enter the event name for setting budget: ").strip().lower()

    # Validate whether the event exists
    event_exists = False
    with open(Event_File, mode='r') as file:
        reader = csv.reader(file)
        for row in reader:
            if row[0].strip().lower() == Event_Name:
                event_exists = True
                break

    if not event_exists:
        print(f"Event '{Event_Name}' not found.")
        return

    # Input budget values with validation
    try:
        venue_budget = float(input("Enter budget for Venue in TK: "))
        catering_budget = float(input("Enter budget for Catering in TK: "))
        entertainment_budget = float(input("Enter budget for Entertainment in TK: "))
        miscellaneous_budget = float(input("Enter budget for Miscellaneous in TK: "))
    except ValueError:
        print("Invalid input. Please enter numeric values for budgets.")
        return

    # Save the budget for the event
    with open(Budget_File, mode='a', newline='') as file:
        writer = csv.writer(file)
        writer.writerow([Event_Name, venue_budget, catering_budget, entertainment_budget, miscellaneous_budget])
    print(f"Budget set successfully for event '{Event_Name}'")

# Function to record expenses for an event
def record_expenses():
    Event_Name = input("Enter the event name to record expenses: ").strip().lower()

    # Validate if the event exists
    event_exists = False
    with open(Event_File, mode='r') as file:
        reader = csv.reader(file)
        for row in reader:
            if row[0].strip().lower() == Event_Name:
                event_exists = True
                break

    if not event_exists:
        print(f"Event '{Event_Name}' not found.")
        return

    # Record expenses with validation
    try:
        venue_expense = float(input("Enter expense for Venue: "))
        catering_expense = float(input("Enter expense for Catering: "))
        entertainment_expense = float(input("Enter expense for Entertainment: "))
        miscellaneous_expense = float(input("Enter expense for Miscellaneous: "))
    except ValueError:
        print("Invalid input. Please enter numeric values for expenses.")
        return

    # Read the budget for the event
    budget_found = False
    with open(Budget_File, mode='r') as file:
        reader = csv.reader(file)
        for row in reader:
            if row[0].strip().lower() == Event_Name:
                venue_budget, catering_budget, entertainment_budget, miscellaneous_budget = map(float, row[1:])
                budget_found = True
                break

    if not budget_found:
        print(f"Budget for event '{Event_Name}' not found.")
        return

    # Calculate totals
    total_expenses = venue_expense + catering_expense + entertainment_expense + miscellaneous_expense
    total_budget = venue_budget + catering_budget + entertainment_budget + miscellaneous_budget
    remaining_budget = total_budget - total_expenses

    # Display results
    print(f"\nTotal Expenses: {total_expenses:.2f}")
    print(f"Total Budget: {total_budget:.2f}")

    if total_expenses > total_budget:
        print("Warning: Expenses exceed the budget!")
    else:
        print(f"Remaining Budget: {remaining_budget:.2f}")

# Function to remove the budget for an event
def remove_event_budget():
    Event_Name = input("Enter the event name to remove its budget or keep blank to keep the previous budget : ").strip().lower()

    # Read all entries from the budget file
    updated_rows = []
    budget_found = False

    with open(Budget_File, mode='r') as file:
        reader = csv.reader(file)
        for row in reader:
            if row[0].strip().lower() == Event_Name:
                budget_found = True
            else:
                updated_rows.append(row)

    if not budget_found:
        print(f"Budget for event '{Event_Name}' not found.")
        return

    # Overwrite the budget file with the remaining entries
    with open(Budget_File, mode='w', newline='') as file:
        writer = csv.writer(file)
        writer.writerows(updated_rows)

    print(f"Budget for event '{Event_Name}' has been removed.")

# Call the functions
initialize_csv()
set_budget()
record_expenses()
remove_event_budget()
```

```python
#5
#to find events by name,date o location
def search_events():
    print("\nSearch Events")
    search_type = input("Search by (1. Name, 2. Date, 3. Location): ").strip()
    query = input("Enter search query: ").strip()

    results_found = False

    with open(Event_File, 'r') as file:
        reader = csv.reader(file)
        for row in reader:
            if (search_type == "1" and query.lower() in row[0].lower()) or \
               (search_type == "2" and query in row[1]) or \
               (search_type == "3" and query.lower() in row[2].lower()):
                print(f"{row[0]} at {row[1]} in {row[2]} with description:{row[3]}")
                results_found = True

    if not results_found:
        print("No events found.")
search_events()

#to send email invitation (question 5)
def send_invitations():
    Event_Name = input("Enter the event name to send invitations: ").strip().lower()
    attendees_found = False

    with open(Attendee_File, 'r') as file:
        reader = csv.reader(file)
        print(f"\nSending Invitations for '{Event_Name}':")
        for row in reader:
            if row[0] == Event_Name:
                attendees_found = True
                print(f"Sending invitation to {row[1]} ({row[2]}) - RSVP Status: {row[3]}")

    if not attendees_found:
        print(f"No attendees found for event '{Event_Name}'.")
send_invitations()


#to display event in a simple calender format(question 5)
import calendar

def display_calendar():
    year = int(input("Enter the year: "))
    month = int(input("Enter the month (1-12): "))

    print("\n" + calendar.month(year, month))
    print("Events in this month:")

    with open(Event_File, 'r') as file:
        reader = csv.reader(file)
        for row in reader:
            try:
                event_date = datetime.strptime(row[1], '%Y-%m-%d %I:%M %p')
                if event_date.year == year and event_date.month == month:
                    print(f"- {event_date.day}: {row[0]} at {row[2]}")
            except ValueError:
              pass

display_calendar()


#to export data
def export_data():
    export_file = input("Enter the name for the export file (e.g., export.csv): ").strip()
    print(f"\nExporting data to: {export_file}")

    try:
        with open(export_file, 'w', newline='') as file:
            writer = csv.writer(file)
            writer.writerow(["Event Name", "Date and Time", "Location", "Description", "Attendee Name", "Attendee Email", "RSVP Status"])

            with open(Event_File, 'r') as events_file:
                event_reader = list(csv.reader(events_file))

            with open(Attendee_File, 'r') as attendees_file:
                attendee_reader = list(csv.reader(attendees_file))

            # Skipping headers in both files
            for event in event_reader[1:]:
                for attendee in attendee_reader[1:]:
                    if event[0] == attendee[0]:  # Match event name
                        print(f"Writing to file: {event} + {attendee[1:]}")
                        writer.writerow(event + attendee[1:])

        print(f"\nData successfully exported to {export_file}.")
    except Exception as e:
        print(f"Error: {e}")
export_data()

```

```python
#6

import csv
from datetime import datetime,timedelta
import pandas as pd
import os

#Define the name of csv file to store event details
Event_File="events.csv"
def initialize_csv():
  with open(Event_File,mode='a',newline='' ) as file:
    if file.tell()==0:
     csv.writer(file).writerow(["Event Name","Date and Time","Location","Description"])

# #Define the name of csv file to store Attendees details
Attendee_File = "attendees.csv"

# Initialize the CSV file
def initialize_csv():
  with open(Attendee_File,'a',newline='') as file:
    if file.tell()==0:
      csv.writer(file).writerow(["Event name","Name","Email","RSVP Status"])

#To create an event
def Create_Event():
  print("\n Create a new event")
  Event_Name=input("Enter the event name:").strip().lower()
  Date_and_Time=input("Enter Date and Time (YYYY-MM-DD HH:MM AM/PM):").strip().lower()
  #[validation of date and time]
  try:
    datetime.strptime(Date_and_Time, '%Y-%m-%d %I:%M %p')
  except ValueError:
    print("Invalid Date-Time format.Please use YYYY-MM-DD HH:MM AM/PM")
    return


  Location=input("Enter location:").strip().lower()
  Description =input("Enter Description:").strip().lower()
                    #To store event details in csv file
  if Event_Name and Date_and_Time and Location and Description:
    with open(Event_File,mode='a',newline='') as file:
      writer=csv.writer(file)
      writer.writerow([Event_Name,Date_and_Time,Location,Description])
    print("Event created successfully ")
  else:
    print("All fields are required to create an event")

#to edit events
def edit_event():
  with open(Event_File, 'r') as file:
        events = list(csv.reader(file))

  if len(events) <= 1:
    print("No events to edit")
    return
  print("\n Here the list of all available events")
  for i, event in enumerate(events[1:], start=1):
    print(f"{i}. {event[0]} - {event[1]} - {event[2]} - {event[3]}")

  event_index = int(input("Enter event number to edit: ")) - 1
  if 0 <= event_index < len(events) - 1:
    event = events[event_index + 1]

  #now we edit the details
    event[0]=input("enter the new event name or keep blank to skip:") or event[0]
    event[1]=input("enter the new Date and Time or keep blank to skip:") or event[1]
    event[2]=input("enter the new location or keep blank to skip:") or event[2]
    event[3]=input("enter the new description or keep blank to skip:") or event[3]

#to save the update
    with open(Event_File, 'w', newline='') as file:
      csv.writer(file).writerows(events)

    print("Event updated successfully")
  else:
    print("Invalid event number")

# to delete event
def delete_event():
#to read events
  with open(Event_File, 'r') as file:
    events = list(csv.reader(file))

  if len(events) <=1:
    print("No events to delete")
    return
  print("\n Here the list of all available events")
  for i,event in enumerate(events[1:], start=1):
    print(f"{i}. {event[0]} - {event[1]} - {event[2]} - {event[3]}")

  event_index=int(input("Enter event number to delete:")) -1

  if 0 <= event_index < len(events) - 1:
    del events[event_index + 1]
    print("Event deleted successfully")
  else:
    print("Invalid event number")

  #to save the update
  with open(Event_File, 'w', newline='') as file:
    csv.writer(file).writerows(events)

# to view events
def view_event():
  with open(Event_File, 'r') as file:
        events = list(csv.reader(file))

  if len(events) <= 1:
    print("No events to view")
    return
  print("\n Here the list of all available events")
  for i, event in enumerate(events[1:], start=1):
    print(f"{i}. {event[0]} - {event[1]} - {event[2]} - {event[3]}")


# To add an attendee to an event
def add_attendee():
    Event_Name = input("Enter the event name: ").strip().lower()
    # Validate if the event exists
    event_exists = False
    with open(Event_File, mode='r') as file:
        reader = csv.reader(file)
        for row in reader:
            if row[0] == Event_Name:
                event_exists = True
                break

    if not event_exists:
        print(f"Event '{Event_Name}' not found. Please create the event first.")
        return

    Name = input("Attendee Name: ").strip().lower()
    Email = input("Attendee Email(e.g. abcd@gmail.com): ")
    if "@" not in Email and "." not in Email:
      print("Invalid Email. Please try again.")
      return
    RSVP_Status = input("RSVP Status (Confirmed/Declined/Maybe): ").strip().lower()
    if RSVP_Status not in ["Confirmed","Declined","Maybe"]:
      print("Invalid RSVP Status.Try Again")
      return

#To store attendee details in csv file
    if Event_Name and Name and Email and RSVP_Status:
      with open(Attendee_File, mode='a', newline='') as file:
        writer=csv.writer(file)
        writer.writerow([Event_Name,Name,Email,RSVP_Status])
      print(f"Attendee '{Name}' added to '{Event_Name}'.")
    else:
      print("All fields are required.")



# To view attendees for a specific event(###2)
def view_attendees():
    Event_Name = input("Enter the event name to view attendees: ").strip().lower()
    attendees_found = False
    with open(Attendee_File, 'r') as file:
        reader = csv.reader(file)
        for row in reader:
            if row[0] == Event_Name:
                if not attendees_found:
                    print(f"\nAttendees for event '{Event_Name}':")
                    attendees_found = True
                print(f"Name: {row[1]}, Email: {row[2]}, RSVP Status: {row[3]}")

    if not attendees_found:
        print(f"No attendees found for event '{Event_Name}'.")


#To remove attendees for each event(######2)
def remove_attendee():
#to read events
  with open(Attendee_File, 'r') as file:
    attendees = list(csv.reader(file))

  if len(attendees) <=1:
    print("No attendee to delete")
    return
  print("\n Here the list of all available attendee details for events")
  for i,attendee in enumerate(attendees[1:], start=1):
    print(f"{i}. {attendee[0]} - {attendee[1]} - {attendee[2]} - {attendee[3]}")

  attendee_index=int(input("Enter attendee number to delete:")) -1

  if 0 <= attendee_index < len(attendees) - 1:
    del attendees[attendee_index + 1]
    print("attendee deleted successfully")
  else:
    print("Invalid event number")

  #to save the update
  with open(Attendee_File, 'w', newline='') as file:
    csv.writer(file).writerows(attendees)


def upcoming_events():
 # to read events
  with open(Event_File, 'r') as file:
    events = list(csv.reader(file))

  if len(events) <= 1:
    print("No upcoming events")
    return

  print("\n Here All Upcoming Events:")
  current_time = datetime.now()
  for event in events[1:]:
    Date_and_Time= datetime.strptime(event[1], '%Y-%m-%d %I:%M %p')
    if Date_and_Time > current_time:
      print(f"{event[0]} - {event[1]} - {event[2]} - {event[3]}")


#to filter events by Date or Location
def Filter_Events():
    filter_by = input("Filter by 'date' or 'location': ").strip().lower()
    filter_value = input(f"Enter the {filter_by}: ").strip()

    with open(Event_File, 'r') as file:
        events = list(csv.reader(file))
    if len(events) <= 1:
        print("No events available to filter.")
        return

    print(f"\nEvents filtered by {filter_by} ({filter_value}):")
    filtered_events = []
    for i, event in enumerate(events[1:], start=1):
        if (filter_by == 'date' and filter_value in event[1]) or (filter_by == 'location' and filter_value in event[2]):
            filtered_events.append(event)
            print(f"{i}. {event[0]} - {event[1]} - {event[2]} - {event[3]}")

    if not filtered_events:
        print(f"No events found matching the criteria '{filter_value}' for '{filter_by}'.")


# Send reminders for events in the next 24 hours
def Send_Reminders():
    with open(Event_File, 'r') as file:
        events = list(csv.reader(file))
    if len(events) <= 1:
        print("No events available for reminders.")
        return

    print("\nReminders for events occurring within the next 24 hours:")
    current_time = datetime.now()
    reminders_sent = False
    for event in events[1:]:
        Date_and_Time = datetime.strptime(event[1], '%Y-%m-%d %I:%M %p')
        if current_time <=Date_and_Time  <= current_time + timedelta(hours=24):
            print(f"Reminder:Event {event[0]} at {event[1]} in {event[2]}. Description: {event[3]}")
            reminders_sent = True

    if not reminders_sent:
        print("No events occurring in the next 24 hours.")


###4

#File path for storing data
Budget_File="budget.csv"
def initialize_csv():
  with open(Budget_File,mode='a',newline='' ) as file:
    if file.tell()==0:
     csv.writer(file).writerow(["Event Name","Venue Budget","Catering Budget","Entertainment Budget","Miscellaneous Budget"])

# Function to set the budget for an event
def set_budget():
    Event_Name = input("Enter the event name for setting budget: ").strip().lower()

    # Validate whether the event exists
    event_exists = False
    with open(Event_File, mode='r') as file:
        reader = csv.reader(file)
        for row in reader:
            if row[0].strip().lower() == Event_Name:
                event_exists = True
                break

    if not event_exists:
        print(f"Event '{Event_Name}' not found.")
        return

    # Input budget values with validation
    try:
        venue_budget = float(input("Enter budget for Venue in TK: "))
        catering_budget = float(input("Enter budget for Catering in TK: "))
        entertainment_budget = float(input("Enter budget for Entertainment in TK: "))
        miscellaneous_budget = float(input("Enter budget for Miscellaneous in TK: "))
    except ValueError:
        print("Invalid input. Please enter numeric values for budgets.")
        return

    # Save the budget for the event
    with open(Budget_File, mode='a', newline='') as file:
        writer = csv.writer(file)
        writer.writerow([Event_Name, venue_budget, catering_budget, entertainment_budget, miscellaneous_budget])
    print(f"Budget set successfully for event '{Event_Name}'")

# Function to record expenses for an event
def record_expenses():
    Event_Name = input("Enter the event name to record expenses: ").strip().lower()

    # Validate if the event exists
    event_exists = False
    with open(Event_File, mode='r') as file:
        reader = csv.reader(file)
        for row in reader:
            if row[0].strip().lower() == Event_Name:
                event_exists = True
                break

    if not event_exists:
        print(f"Event '{Event_Name}' not found.")
        return

    # Record expenses with validation
    try:
        venue_expense = float(input("Enter expense for Venue: "))
        catering_expense = float(input("Enter expense for Catering: "))
        entertainment_expense = float(input("Enter expense for Entertainment: "))
        miscellaneous_expense = float(input("Enter expense for Miscellaneous: "))
    except ValueError:
        print("Invalid input. Please enter numeric values for expenses.")
        return

    # Read the budget for the event
    budget_found = False
    with open(Budget_File, mode='r') as file:
        reader = csv.reader(file)
        for row in reader:
            if row[0].strip().lower() == Event_Name:
                venue_budget, catering_budget, entertainment_budget, miscellaneous_budget = map(float, row[1:])
                budget_found = True
                break

    if not budget_found:
        print(f"Budget for event '{Event_Name}' not found.")
        return

    # Calculate totals
    total_expenses = venue_expense + catering_expense + entertainment_expense + miscellaneous_expense
    total_budget = venue_budget + catering_budget + entertainment_budget + miscellaneous_budget
    remaining_budget = total_budget - total_expenses

    # Display results
    print(f"\nTotal Expenses: {total_expenses:.2f}")
    print(f"Total Budget: {total_budget:.2f}")

    if total_expenses > total_budget:
        print("Warning: Expenses exceed the budget!")
    else:
        print(f"Remaining Budget: {remaining_budget:.2f}")

# Function to remove the budget for an event
def remove_event_budget():
    Event_Name = input("Enter the event name to remove its budget or keep blank to keep the previous budget : ").strip().lower()

    # Read all entries from the budget file
    updated_rows = []
    budget_found = False

    with open(Budget_File, mode='r') as file:
        reader = csv.reader(file)
        for row in reader:
            if row[0].strip().lower() == Event_Name:
                budget_found = True
            else:
                updated_rows.append(row)

    if not budget_found:
        print(f"Budget for event '{Event_Name}' not found.")
        return

    # Overwrite the budget file with the remaining entries
    with open(Budget_File, mode='w', newline='') as file:
        writer = csv.writer(file)
        writer.writerows(updated_rows)

    print(f"Budget for event '{Event_Name}' has been removed.")





def main():
    initialize_csv()
    while True:
        print("\n EventManagement System")
        print("1. Create Event")
        print("2. Edit Event")
        print("3. Delete Event")
        print("4.view Events")
        print("5.Manage attendees")
        print("6.Track Schedule")
        print("7.Budget Management")
        print("8.Exit")
        choice = input("Enter your choice(1-8): ")

        if choice == "1":
            Create_Event()
        elif choice == "2":
            edit_event()
        elif choice == "3":
            delete_event()
        elif choice == "4":
            view_event()
        elif choice == "5":
            print("\nManage attendees")
            while True:
                print("1. Add Attendee")
                print("2. View Attendees")
                print("3. Remove Attendee")
                print("4. Back to Main Menu")
                attendee_choice = input("Enter your choice(1-4): ")

                if attendee_choice == "1":
                    add_attendee()
                elif attendee_choice == "2":
                    view_attendees()
                elif attendee_choice == "3":
                    remove_attendee()
                elif attendee_choice == "4":
                    print("Exiting the management attendee")
                    break
                else:
                    print("Invalid choice.Try again")
        elif choice == "6":
            print("\nTrack Schedule")
            while True:
                print("1. Upcoming Events")
                print("2. Filter Events")
                print("3. Send Reminders")
                print("4. Back to Main Menu")
                schedule_choice = input("Enter your choice(1-4): ")
                if schedule_choice == "1":
                    upcoming_events()
                elif schedule_choice == "2":
                    Filter_Events()
                elif schedule_choice == "3":
                    Send_Reminders()
                elif schedule_choice == "4":
                    break
                else:
                    print("Invalid choice.Try again")
        elif choice == "7":
            print("\nBudget Management")
            set_budget()
            record_expenses()
            remove_event_budget()
        elif choice == "8":
              print("Exiting the program. Thank You")
              break
        else:
            print("Invalid choice.Try again")

if __name__ == "__main__":
    main()
```
